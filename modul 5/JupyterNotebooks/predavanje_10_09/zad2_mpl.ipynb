```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import pandas as pd
covid_df = pd.read_csv("casedistribution.csv", parse_dates=["dateRep"], dayfirst=True)
covid_df
```

```python
plt.plot(covid_df.dateRep, covid_df.cases)
```

```python
var_graf = covid_df.groupby("dateRep", as_index=False).mean(numeric_only=False)
plt.plot(var_graf.dateRep, var_graf.cases)

```

```python
#[marker][stil_linije][boja] 
##  https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html 
plt.plot(var_graf.dateRep, covid_df.cases, "ob")
```
