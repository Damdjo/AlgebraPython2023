Igra na sreću s dvije šesterostrane kockice.
Pravila: igrač baca dvije šesterostrane kockice. 
Ako je zbroj dobivenih brojeva u prvom bacanju 7 ili 11, igrač je pobijedio. 
Ako je zbroj dobivenih brojeva: 4, 5, 6, 8, 9 ili 10, onda taj broj postaje CILJ, odnosno igrač mora bacati kockice sve dok zbroj dobivenih brojeva ne bude CILJ. 
Ako je u prvom ili bilo kojem drugom bacanju igrač dobio zbroj brojeva: 2, 3 ili 16, igrač je izgubio.Igra na sreću s dvije šesterostrane kockice.          

```python
import random
#bacanje + checkwin
def throw_die(no_die_sides:int=6, no_throws:int=1, no_dies:int=1):
    total = []
    for throw in range(no_throws):
        throws_list = []
        for kockica in range(no_dies):
            roll = random.randint(1,no_die_sides)
            throws_list.append(roll)
        total.append(tuple(throws_list))
    return total

def check_throw(roll_tuple, counter = 1,goal = 0):
    total = sum(roll_tuple)
    
    if goal == 0 and counter == 1:
        match total: # znak "|" je or operator
            case 7 | 11:
                return("Win")

            case 2 | 3 | 12:
                return("Lose")

            case 4 | 5 | 6 | 8 | 9 | 10:
                print("test")
                return total

            case _:
                return("Throw again")
            
    elif goal != 0 and counter > 1:
        if total == goal:
            return("Win")
        else:            
            match total: # znak "|" je or operator

                case 2 | 3 | 12:
                    return("Lose")
                case _:                
                    return("Throw again")
            
        
                
            
            

```

```python
#prikaz rezultata
def results(throw_tuple, counter:int=1, goal= 0):
    for index, die in enumerate(throw_tuple):
        
        
        check = check_throw(throw_tuple[index], counter, goal)
        if goal == 0 and type(check) == int and counter == 1:
            goal = check
            check = f"Goal is now {goal}" 
        else:
            goal = goal
        
        print(f"""
        Throw no. {counter+1}:
            You rolled numbers {die[0]} and {die[1]}!
            Sum of rolled numbers {sum(throw_tuple[index])}        
            {check}
            """)
        match check:
            case "Win" | "Lose":
                return 0, 0
            case _:
                return 1, goal
            
```

```python
#main

counter = 1
goal = 0
status = 1
while status != 0:    
    throw = throw_die(6,1,2)
    status, goal = results(throw, counter, goal)
    
    counter+=1
    #if status !=0:
        #input("Press enter to throw again...")
    
```
