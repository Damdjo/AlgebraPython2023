```python
import pandas as pd

listings_df = pd.read_csv("airbnb_new_york_city_listings.csv")
```

```python
listings_df
```

```python
listings_df.dtypes
```

```python
listings_df["price"] = listings_df["price"].astype("float64")
listings_df["last_review"] = pd.to_datetime(listings_df["last_review"])
```

```python
listings_df["last_review_year"] = listings_df["last_review"].dt.year
listings_df["last_review_year"] = pd.to_numeric(listings_df["last_review_year"])
```

```python
#name "No Name Listing"
#host_name "J.Doe"
#reviews_per_month 0.00
#last_review_year 0
#last_review 01-01-1970

###  https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html

#fillna()
#replace()
#interpolate()
```

```python
listings_df["name"].fillna("No Name Listing", inplace = True)
listings_df["host_name"].fillna("J.Doe", inplace = True)
listings_df["reviews_per_month"].fillna(0.00, inplace = True)
listings_df["last_review_year"].fillna(0, inplace = True)
listings_df["last_review"].fillna("01-01-1970", inplace = True)
```

```python
listings_df.dtypes
```

```python
listings_df["last_review_year"] = listings_df["last_review_year"].astype("int64")
```

```python
listings_df.dtypes
```

```python
listings_df["name"] = listings_df["name"].str.strip()
listings_df["host_name"] = listings_df["host_name"].str.strip()
listings_df["neighbourhood_group"] = listings_df["neighbourhood_group"].str.strip()
listings_df["neighbourhood"] = listings_df["neighbourhood"].str.strip()
listings_df["room_type"] = listings_df["room_type"].str.strip()
```
