```python
okvir = plt.figure(figsize=(8,8))
gs = okvir.add_gridspec(3,3)

gore_lijevo = okvir.add_subplot(gs[0,0]) #prvi red, prvi stupac
gore_sredina = okvir.add_subplot(gs[1,0]) #prvi red, drugi stupac
gore_desno = okvir.add_subplot(gs[:2,1:]) #prva 2 reda i stupci od indeksa 1 nadalje

dolje = okvir.add_subplot(gs[2,:]) #treci red, svi stupci
```

Promjenom parametara metoda add_gridspec() i add_subplot(), kreirajte gornji raspored okvira kao u ogledalu.
Promjenom parametara metoda add_gridspec() i add_subplot(), kreirajte raspored okvira tako da je jedan okvir unutar svake Ä‡elije.
Promjenom parametara metoda add_gridspec() i add_subplot(), kreirajte raspored okvira tako da je jedan okvir unutar svakog reda.
Promjenom parametara metoda add_gridspec() i add_subplot(), kreirajte raspored okvira tako da je jedan okvir unutar svakog stupca. 

```python
#ogledalo
okvir = plt.figure(figsize=(8,8))
gs = okvir.add_gridspec(3,3)

gore_desno = okvir.add_subplot(gs[0,2]) #prvi red, treci stupac
sredina_desno = okvir.add_subplot(gs[1,2]) #drugi red, treci stupac
gore_lijevo = okvir.add_subplot(gs[:2,0:2]) #prva 2 reda i stupci od indeksa 0 do 2

dolje = okvir.add_subplot(gs[2,:]) #treci red, svi stupci
```

```python
#samo celije
okvir = plt.figure(figsize=(8,8))
gs = okvir.add_gridspec(3,3)

gore_lijevo = okvir.add_subplot(gs[0,0]) #prvi red, prvi stupac
gore_sredina = okvir.add_subplot(gs[0,1]) #prvi red, drugi stupac
gore_desno = okvir.add_subplot(gs[0,2]) #prvi red, treci stupac

sredina_lijevo = okvir.add_subplot(gs[1,0]) #drugi red, prvi stupac
sredina_sredina = okvir.add_subplot(gs[1,1]) #drugi red, drugi stupac
sredina_desno = okvir.add_subplot(gs[1,2]) #drugi red, treci stupac

dolje_lijevo = okvir.add_subplot(gs[2,0]) #treci red, prvi stupac
dolje_sredina = okvir.add_subplot(gs[2,1]) #treci red, drugi stupac
dolje_desno = okvir.add_subplot(gs[2,2]) #treci red, treci stupac
```

```python
#puni redovi
okvir = plt.figure(figsize=(8,8))
gs = okvir.add_gridspec(3,3)

gore = okvir.add_subplot(gs[0,:]) #prvi red, svi stupci
sredina = okvir.add_subplot(gs[1,:]) #prvi red, svi stupci
dolje = okvir.add_subplot(gs[2,:]) #prvi red, svi stupci
```

```python
#samo stupci
okvir = plt.figure(figsize=(8,8))
gs = okvir.add_gridspec(3,3)

lijevo = okvir.add_subplot(gs[:,0]) #svi redovi, prvi stupac
sredina = okvir.add_subplot(gs[:,1]) #svi redovi, drugi stupac
desno = okvir.add_subplot(gs[:,2]) #svi redovi, treci stupac
```

```python
import matplotlib.dates as mdates
covid_df["dateRep"] = mdates.date2num(covid_df["dateRep"])
covid_df.plot()
```

```python
covid_df.plot(
    kind= "line",
    x= "dateRep",
    y= "cases",
    figsize= (10,5),
    style= "-b",
    legend = True,
    title = "Broj covid sluÄajeva"

)
```

```python
covid_df.groupby("dateRep", as_index = False).mean(numeric_only=True).tail(40).plot(
    kind= "line",
    x= "dateRep",
    y= ["cases","deaths"],
    #xlabel="Datum",
    #ylabel="Broj zaraÅ¾enih",
    subplots= True,
    layout= (2,1), #2 retka, 1 stupac
    figsize= (18,8),
    color= ("green","red"),
    linestyle= "solid",
    legend = True,
    title = "Broj covid sluÄajeva"
)
```

```python
novi_slucajevi_prosjek = covid_df.pivot_table(
    index = covid_df.index,
    columns= "countriesAndTerritories",
    values= "cases").rolling(7).mean(numeric_only=True)

okvir, axes = plt.subplots(1,3, figsize= (15,5))

novi_slucajevi_prosjek[["Germany"]].plot(
    ax= axes[0],
    style = "-.c")

novi_slucajevi_prosjek[["France"]].plot(
    ax= axes[1],
    style = ["-","--"],
    title = "7 dnevni prosjek novih COVID19 slucajeva"
    )

novi_slucajevi_prosjek[["Croatia"]].plot(
    ax= axes[2],
    style = ["--",":","-"])
```

```python
earthquakes_df = pd.read_csv("earthquakes.csv")
earthquakes_df
```

```python
earthquakes_df[["mag","magType"]].groupby("magType", as_index=False).boxplot(figsize=(15,8),subplots=False)
plt.title("Magnitude zemljotresa po tipu magnitude")
plt.ylabel("Magnituda")
```

```python
#earthquakes_df["placeb"]
earthquakes_df.place.value_counts().iloc[15::-1,].plot(
    kind= "barh",
    figsize= (10,5),
    title= "15 mjesta s najviÅ¡e zemljotresa"
)

plt.xlabel("Zemljotresi")
```

```python
earthquakes_df.groupby("place").tsunami.sum().sort_values().iloc[-15:,].plot(
    kind= "barh",
    figsize= (10,5),
    title= "15 mjesta s najviÅ¡e tsunamija"
)

plt.xlabel("Zemljotresi")
```
