```python
#ogledalo
okvir = plt.figure(figsize=(8,8))
gs = okvir.add_gridspec(3,3)

gore_desno = okvir.add_subplot(gs[0,2]) #prvi red, treci stupac
sredina_desno = okvir.add_subplot(gs[1,2]) #drugi red, treci stupac
gore_lijevo = okvir.add_subplot(gs[:2,0:2]) #prva 2 reda i stupci od indeksa 0 do 2

dolje = okvir.add_subplot(gs[2,:]) #treci red, svi stupci
```

```python
#samo celije
okvir = plt.figure(figsize=(8,8))
gs = okvir.add_gridspec(3,3)

gore_lijevo = okvir.add_subplot(gs[0,0]) #prvi red, prvi stupac
gore_sredina = okvir.add_subplot(gs[0,1]) #prvi red, drugi stupac
gore_desno = okvir.add_subplot(gs[0,2]) #prvi red, treci stupac

sredina_lijevo = okvir.add_subplot(gs[1,0]) #drugi red, prvi stupac
sredina_sredina = okvir.add_subplot(gs[1,1]) #drugi red, drugi stupac
sredina_desno = okvir.add_subplot(gs[1,2]) #drugi red, treci stupac

dolje_lijevo = okvir.add_subplot(gs[2,0]) #treci red, prvi stupac
dolje_sredina = okvir.add_subplot(gs[2,1]) #treci red, drugi stupac
dolje_desno = okvir.add_subplot(gs[2,2]) #treci red, treci stupac
```

```python
#puni redovi
okvir = plt.figure(figsize=(8,8))
gs = okvir.add_gridspec(3,3)

gore = okvir.add_subplot(gs[0,:]) #prvi red, svi stupci
sredina = okvir.add_subplot(gs[1,:]) #prvi red, svi stupci
dolje = okvir.add_subplot(gs[2,:]) #prvi red, svi stupci
```
